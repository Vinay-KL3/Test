```python
import re
from datetime import datetime, timedelta

def find_and_convert_dates(text):
    # Regular expression pattern for matching different date formats
    date_pattern = r'\b(\d{1,2}[-/]\d{1,2}[-/]\d{4})\b|\b(\b(?:in next|tomorrow|today)\s*(\d+)?\s*(day|days|week|weeks|month|months|year|years)\b)|\b(?:Monday|Tuesday|Wednesday|Thursday|Friday|Saturday|Sunday)\b|\b(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)\s*\d{1,2}(?:st|nd|rd|th)?\b'
    
    # Find all occurrences of the date pattern in the text
    date_matches = re.finditer(date_pattern, text, re.IGNORECASE)
    
    # Get today's date
    today = datetime.today().date()

    dates = []
    for match in date_matches:
        date_str = match.group()
        
        # If the date is in the format DD-MM-YYYY or DD/MM/YYYY
        if re.match(r'\d{1,2}[-/]\d{1,2}[-/]\d{4}', date_str):
            date_obj = datetime.strptime(date_str, '%d-%m-%Y').date() if '-' in date_str else datetime.strptime(date_str, '%d/%m/%Y').date()

        # If the date is a relative date (e.g., in next 2 days, tomorrow, today)
        else:
            if 'next' in date_str.lower():
                interval = int(match.group(3) or '1')
                if 'day' in date_str:
                    date_obj = today + timedelta(days=interval)
                elif 'week' in date_str:
                    date_obj = today + timedelta(weeks=interval)
                elif 'month' in date_str:
                    date_obj = today.replace(month=today.month + interval)
                elif 'year' in date_str:
                    date_obj = today.replace(year=today.year + interval)
            elif 'tomorrow' in date_str.lower():
                date_obj = today + timedelta(days=1)
            elif 'today' in date_str.lower():
                date_obj = today
            else:
                date_obj = datetime.strptime(date_str, '%b %d').date()

        dates.append(date_obj)

    return dates

# Example text
sample_text = "The event is scheduled on 15-07-2023 and in next 2 days."

# Find and convert dates in the text
dates_found = find_and_convert_dates(sample_text)
print(dates_found)

# Print the relative dates compared to today's date
for date_obj in dates_found:
    difference = (date_obj - datetime.today().date()).days
    if difference == 0:
        print(f"Today")
    elif difference == 1:
        print(f"Tomorrow")
    elif difference > 1:
        print(f"In {difference} days")
    else:
        print(f"{-difference} days ago")
```
