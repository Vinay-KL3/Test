```python
import spacy

# Load the pre-trained SpaCy NER model (e.g., "en_core_web_sm" for English)
nlp = spacy.load("en_core_web_sm")

# Example text
text = "Apple Inc. was founded by Steve Jobs and Steve Wozniak on April 1, 1976, in California."

# Process the text using the NER model
processed_text = nlp(text)

# Print the entities found in the text
for ent in processed_text.ents:
    if ent.label_ == "DATE":
#     print(ent.text, "   |   ", ent.label_, )
        print(ent.text)
```
