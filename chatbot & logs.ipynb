```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display

# Create a DataFrame to store chat logs
chat_logs = pd.DataFrame(columns=['Session ID', 'User Input', 'Bot Response'])

# Generate a unique session ID for each chat session
session_id = 1

# Global variables to store user selections
selected_date = None
selected_doctor = None

# Function to handle user input and bot response
def handle_chat(user_input):
    global session_id, selected_date, selected_doctor

    # Update user selections
    if 'Date' in user_input:
        selected_date = user_input['Date']
    if 'Doctor' in user_input:
        selected_doctor = user_input['Doctor']
    
    # Check if both date and doctor are selected
    if selected_date and selected_doctor:
        # Bot's response when both date and doctor are selected
        bot_response = process_user_input()
        
        # Display the bot's response
        display_bot_response(bot_response)
        
        # Log the chat in the DataFrame
        chat_logs.loc[len(chat_logs)] = [session_id, user_input, bot_response]
        
        # Reset user selections for the next chat session
        selected_date = None
        selected_doctor = None
        session_id += 1

# Function to process user input and return bot's response
def process_user_input():
    # Implement your logic to process the user selections and generate the bot's response
    # Customize this function based on your appointment booking logic
    return f"Your appointment is booked for {selected_date} with Dr. {selected_doctor}."

# Function to display the bot's response
def display_bot_response(response):
    print(f"Bot: {response}\n")

# Create button options for date and doctor
date_options = ['2023-06-18', '2023-06-19', '2023-06-20']
doctor_options = ['Dr. Smith', 'Dr. Johnson', 'Dr. Brown']

# Create the buttons
date_buttons = [widgets.Button(description=date) for date in date_options]
doctor_buttons = [widgets.Button(description=doctor) for doctor in doctor_options]

# Create a container for the buttons
button_container = widgets.HBox(date_buttons + doctor_buttons)

# Register button click event handlers
for button in date_buttons + doctor_buttons:
    button.on_click(lambda button: handle_chat({'Date': button.description}) if button in date_buttons
                    else handle_chat({'Doctor': button.description}))

# Display the button container
display(button_container)
```

```python
import streamlit as st
from chatterbot import ChatBot
from chatterbot.trainers import ChatterBotCorpusTrainer

# Create a ChatBot instance
chatbot = ChatBot('MyChatBot')
trainer = ChatterBotCorpusTrainer(chatbot)

# Train the chatbot using an English corpus
trainer.train('chatterbot.corpus.english')

# Define the Streamlit app
def main():
    st.title("Chatbot App")

    # Add a button to open the chatbot
    if st.button("Open Chatbot"):
        # Get user input using an input box
        user_input = st.text_input("You:", key="user_input")

        # Chatbot response
        bot_response = chatbot.get_response(user_input)

        # Display the chatbot response
        st.text_area("Chatbot:", value=str(bot_response), key="bot_response")

# Run the Streamlit app
if __name__ == "__main__":
    main()
```
