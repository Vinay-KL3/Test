```python
#import all dependencies

import speech_recognition as sr
import pyttsx3
import datetime
import pywhatkit
import wikipedia
import webbrowser
import imdb
import pprint

moviesdb = imdb.IMDb()
```

```python
#define a function for talking

machine = pyttsx3.init()
machine.setProperty('rate', 70)
machine.setProperty('pitch', 1.25)

def talk(text):
    machine.say(text)
    machine.runAndWait()    
```

```python
#define a function for listening

listener = sr.Recognizer()

def give_instruction():
    
    #global instruction  #making this a global variable
    
    
    with sr.Microphone() as origin:
        print("am listening, pleask ask me something...")
        speech = listener.listen(origin)
        instruction = listener.recognize_google(speech)

        instruction = instruction.lower()
        if "alexa" in instruction:
            instruction = instruction.replace("alexa","")
            print(instruction)
        return instruction
```

```python
#define if conditions for virtual bot

def Alexa():
    instruction = give_instruction()
    print(instruction)
    
    
    if "how are you" in instruction:
        response = "am good, thanks for asking. how are you?"
        print(response)
        talk(response)
        
        
    elif "play" in instruction:
        song = instruction.replace("play", "")
        response = "playing" + song
        print(response)
        talk(response)
        pywhatkit.playonyt(song)
        
        
    elif "what is" in instruction:
        item = instruction.replace("what is", "")
        response = "checking for" + item
        print(response)
        talk(response)
        response = wikipedia.summary(item, sentences=5)
        print(response)
        talk(response)   
        
        
    elif "who is" in instruction:
        item = instruction.replace("who is", "")
        response = "checking for " + item
        print(response)
        talk(response)
        response = wikipedia.summary(item, sentences=5)
        print("according to wikipedia." + response)
        talk("according to wikipedia, " + response)    
        
        
    elif "browse" in instruction:
        item = instruction.replace("browse", "")
        response = "browsing for " + item
        print(response)
        talk(response)
        response = webbrowser.open(item)
        print(response)
        talk(response)  
        
        
    elif "movie" in instruction:
        response = "browsing for " + instruction
        print(response)
        talk(response)
        response = moviesdb.search_movie(instruction)
        print(response)
        talk(response)          
        
    elif "bye" in instruction:
        talk("okay, bye!") 
        return False
    
    
    
    elif "hello" in instruction:
        response = "Hello! Am i speaking with customer name?"
        print(response)
        talk(response)
        
        
    elif "yes" in instruction:
        response = "Your I C I C I credit card payment of 4457 rupees is pending from 4 days, please make the payment immediately to avoid the impact on your credit score. Thank you! Have a nice day!"
        print(response)
        talk(response)
        return False 
        
        
    else:
        response = "sorry i didnt get it, can you please rephrase and ask again"
        print(response)
        talk(response)
    return True
    
```

```python
#run and use Maaya bot
run = True 
while run == True:
    run = Alexa()
    
# Alexa()   
```
