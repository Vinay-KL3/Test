```python
with open(dataCSV, 'r') as f:
    reader = csv.reader(f)
    the_whole_file = list(reader)    
    
count = 0
while(count < len(the_whole_file)):
    count = count + 1
    TestCaseNo = the_whole_file[count][0]
    Message = the_whole_file[count][3]
    Category = the_whole_file[count][1]
    
    #print(TestCaseNo)
    #print(Message)
    
    
    
    
    if Category == "CBN":
        try:

            CBN(TestCaseNo,Message)
            time.sleep(0.5)

        except Exception as e:
            print(e)
            
            
            
        
    elif Category == "RPN":
        try:

            RPN(TestCaseNo,Message)
            time.sleep(0.5)

        except Exception as e:
            print(e)
                     
                
               
            
            
    elif Category == "RPC":
        try:

            RPC(TestCaseNo,Message)
            time.sleep(0.5)

        except Exception as e:
            print(e)
            
            
            
            
            
    elif Category == "RPC_L":
        try:

            RPC_L(TestCaseNo,Message)
            time.sleep(0.5)

        except Exception as e:
            print(e)
            
            
            
            
            
    elif Category == "RPC_L2":
        try:

            RPC_L2(TestCaseNo,Message)
            time.sleep(0.5)

        except Exception as e:
            print(e)
            
            
            
            
            
    elif Category == "PIC_ASK_QUERIES":
        try:

            PIC_ASK_QUERIES(TestCaseNo,Message)
            time.sleep(0.5)

        except Exception as e:
            print(e)
            
            
            
            
            
    elif Category == "PIC_ASK_AVAILABLE_PAYMENT_METHODS":
        try:

            PIC_ASK_AVAILABLE_PAYMENT_METHODS(TestCaseNo,Message)
            time.sleep(0.5)

        except Exception as e:
            print(e)
                        
                
                
                
      
    elif Category == "PIC_GENERAL":
        try:

            PIC_GENERAL(TestCaseNo,Message)
            time.sleep(0.5)

        except Exception as e:
            print(e)
            
            
            
            
            
            
    elif Category == "PIC_APTP":
        try:

            PIC_APTP(TestCaseNo,Message)
            time.sleep(0.5)

        except Exception as e:
            print(e)
            
            
            
            
            
    else:
        try:

            DELAY_DELAY_REASONS(TestCaseNo,Message)
            time.sleep(0.5)

        except Exception as e:
            print(e)
            
            
            
    

                        
```
