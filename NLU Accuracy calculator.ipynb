```python
import requests
import pandas as pd
from jiwer import wer
```

```python
PATH = "/Users/vinay_kl/Downloads/temptemp.csv"
path = r "/Users/vinay_kl/Downloads/temptemp_out.csv"
```

```python
df = pd.read_csv(PATH)
df.head(2)
```

```python
for index, row in df.iterrows():
        
    intent = row["intent"]
    intent_pred = row["intent_pred"]

    sub_intent = row["sub-intent"]
    sub_intent_pred = row["sub-intent_pred"]
    
    context = row["context"]
    context_pred = row["context_pred"]
    
    sub_context = row["sub-context"]
    sub_context_pred = row["sub-context_pred"]
    
    signal = row["signal"]
    signal_pred = row["signal_pred"]
    
    delay_reason = row["delay_reason"]
    delay_reason_pred = row["delay_reason_pred"]
    
    third_person = row["third_person"]
    third_person_pred = row["third_person_pred"]
    
    humiliate = row["humiliate"]
    humiliate_pred = row["humiliate_pred"]
    
    sentiment = row["sentiment"]
    sentiment_pred = row["sentiment_pred"]
    
    
    intent_accuracy = wer(intent,intent_pred)
    sub_intent_accuracy = wer(sub_intent,sub_intent_pred)
    context_accuracy = wer(context,context_pred)
    sub_context_accuracy = wer(sub_context,sub_context_pred)
    signal_accuracy = wer(signal,signal_pred)
    delay_reason_accuracy = wer(delay_reason,delay_reason_pred)
    third_person_accuracy = wer(third_person,third_person_pred)
    humiliate_accuracy = wer(humiliate,humiliate_pred)
    sentiment_accuracy = wer (sentiment,sentiment_pred)
    
    

    e = (intent_accuracy,'\t',sub_intent_accuracy,'\t',context_accuracy,'\t',sub_context_accuracy,'\t',signal_accuracy,'\t',delay_reason_accuracy,'\t',third_person_accuracy,'\t',humiliate_accuracy,'\t',sentiment_accuracy)

    #print(type(e))
    
    with open(path, 'w') as f:   
        f.write(str(e))
        f.close
```
