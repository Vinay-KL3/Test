```python
import requests
import json
import pandas as pd
from pprint import pprint
import csv
import time
import random
import datetime
import re
```

```python
EndPoint = "http://216.48.184.125:8090/predict_number"
dataCSV = "/Users/vinay_kl/Downloads/Digits to Numbers - Sheet1.csv"
```

```python
def D2N(text = None):
    headers = {'Content-Type':'application/json'}    
    payload = json.dumps ({
        "text": text,
        "language": "odia"
    })
    response = requests.request("POST", EndPoint, headers=headers, data=payload).json()

    print(response)
```

```python
with open(dataCSV, 'r') as f:
    reader = csv.reader(f)
    the_whole_file = list(reader)    
    
count = 0
while(count < (len(the_whole_file)-1)):
    count = count + 1
    text = the_whole_file[count][0]
#     print(text)
    
    try:
        D2N(text = text)
        #time.sleep(0.5)
    except Exception as e:
        print(e)
    
    
    
    
        
    
```
