```python
import os
import csv
#Rename Function
def rename(filePath,audioPath):
    #function: replace audio file name with audio IDs of Primary (unique) responses
    #audioPath = path/location of the the unique/primary response
    #filePath = template of unique/primary response
    countAudio=1
    source=''
    desti=''
    with open(filePath) as csv_file:
        csv_reader = csv.reader(csv_file)
        for col in csv_reader:
            audioID= col[0] #audio id
            responseID = col[1] #responseID as fileName
            #gender = col [3]
            source=audioPath+responseID+".wav"
            desti=audioPath+audioID+".wav"
            if os.path.exists(source)==True:
                os.rename(source,desti)
                print(responseID+".wav", ' is renamed as:')
                print(audioID+".wav")
                countAudio+=1
                print('audio count is: ',responseID)
            else:
                print("Not Exist:",source)
#template of unique/primary response
temPath='/Users/vinay_kl/Downloads/hin_final_split.csv'
#path/location of the the unique/primary response audios
uniAudioPath='/Users/vinay_kl/Downloads/maia_enhancements_all_2/hindi/entity/'
rename(temPath,uniAudioPath)
```
