```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data for the two lists
list1 = np.random.normal(5, 2, 1000)
list2 = np.random.normal(10, 3, 1000)

# Plotting the histograms
plt.hist(list1, bins=30, density=True, alpha=0.5, label='List 1')
plt.hist(list2, bins=30, density=True, alpha=0.5, label='List 2')

# Calculating the mean and standard deviation for each list
mean1, std1 = np.mean(list1), np.std(list1)
mean2, std2 = np.mean(list2), np.std(list2)

# Create x-axis values
x = np.linspace(min(list1.min(), list2.min()), max(list1.max(), list2.max()), 100)

# Generate the bell curves using the mean and standard deviation
curve1 = np.exp(-(x - mean1)**2 / (2 * std1**2)) / (std1 * np.sqrt(2 * np.pi))
curve2 = np.exp(-(x - mean2)**2 / (2 * std2**2)) / (std2 * np.sqrt(2 * np.pi))

# Plotting the bell curves
plt.plot(x, curve1, color='r', label='Bell curve 1')
plt.plot(x, curve2, color='b', label='Bell curve 2')

# Add legend and labels to the plot
plt.legend()
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Bell Curves')

# Display the plot
plt.show()
```
