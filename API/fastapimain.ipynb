```python
import pandas as pd

df = pd.read_csv("/Users/vinay_kl/Downloads/Fast Text Test - Copy of Stellantis.csv") #, sep="\t", error_bad_lines=False)

print(df.shape)


df.dropna(inplace=True)
print(df.shape)

# df["Final"] = "__label__" +df["intent"] + " " + "__label__" +df["sub_intent"] + " " + "__label__" +df["context"] + " " + "__label__" +df["sub_context"] + " " + "__label__" +df["signal"] + " " + "__label__" +df["reason"] + " " + "__label__" +df["third_party"] + " " + "__label__" +df["humiliate"] + " " + "__label__" +df["sentiment"] + " " + df["Utterance"].astype(str)
# df["Final"] = "__label__" +df["intent"] + " " + df["Utterance"].astype(str)
# df["Final"] = "__label__" + df["Language"] + " " + df["Movie Name"].astype(str)
df["Final"] = "__label__"  +df["intent"]  +"_"+df["signal"]  +"_"+df["reason"]   +" "+   df["Utterance"].astype(str)


# for i in df["Final"]:
#     print(i)

def preprocess(text=None):
    import re
    text = re.sub(r"[^\w\s\']", " ", text)
    text = re.sub(r" +", " ", text) 
    text = text.strip().lower()
    return text



df["Final"] = df["Final"].map(preprocess)

for i in df["Final"]:
    print(i)
    
    
    
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.2)

train.shape, test.shape



train.to_csv("/Users/vinay_kl/Downloads/data_train.csv", columns=["Final"], index=False, header=False)
test.to_csv("/Users/vinay_kl/Downloads/data_test.csv", columns=["Final"], index=False, header=False)


import fasttext

model = fasttext.train_supervised(input = "/Users/vinay_kl/Downloads/data_train.csv", wordNgrams = 6, epoch = 10000)
model.test("/Users/vinay_kl/Downloads/data_train.csv")
```

```python




response = model.predict("i will choose ahmedabad")
# model.get_labels()

A = response[0][0]
A = A.replace("__label__", "")

print(A,"\t",response[1][0])



```
