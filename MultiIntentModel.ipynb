```python
#importrequired csv

import pandas as pd
import fasttext

Stellantis = "/Users/vinay_kl/Downloads/Fast Text Test - Copy of Stellantis.csv"
Collections = "/Users/vinay_kl/Downloads/Fast Text Test - Collections.csv"
CollectionsMulti4 = "/Users/vinay_kl/Downloads/Fast Text Test - Collections_Multi_Lingual (1).csv"

df = pd.read_csv(Collections) #, sep="\t", error_bad_lines=False)

print(df.shape)
df.head(3)
```

```python
def predict(Text):

    A = intent.predict(Text)
    B = sub_intent.predict(Text)
    C = context.predict(Text)
    D = sub_context.predict(Text)
    E = signal.predict(Text)
    F = reason.predict(Text)
    G = third_party.predict(Text)
    H = humiliate.predict(Text)
    I = sentiment.predict(Text)


    def Final(label = None):
        clean = label[0][0]
        clean = clean.replace("__label__", "")
        print(clean,"\t",label[1][0])


    Final(A)
    Final(B)
    Final(C)    
    Final(D)    
    Final(E)    
    Final(F)    
    Final(G)    
    Final(H)    
    Final(I)    
```

```python
#api


from fastapi import FastAPI
from pydantic import BaseModel


app = FastAPI()

class intake(BaseModel):
    Text : str
        
        
@app.post("/predict/{Text}")
def predict(Text : str):

    A = intent.predict(Text)
    B = sub_intent.predict(Text)
    C = context.predict(Text)
    D = sub_context.predict(Text)
    E = signal.predict(Text)
    F = reason.predict(Text)
    G = third_party.predict(Text)
    H = humiliate.predict(Text)
    I = sentiment.predict(Text)


    def Final(label = None):
        clean = label[0][0]
        clean = clean.replace("__label__", "")
        print(clean,"\t",label[1][0])


    Final(A)
    Final(B)
    Final(C)    
    Final(D)    
    Final(E)    
    Final(F)    
    Final(G)    
    Final(H)    
    Final(I)     
```
