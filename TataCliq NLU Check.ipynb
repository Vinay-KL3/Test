```python
import requests
import json
import pandas as pd
from pprint import pprint
import csv
import time
import random
```

```python
dataCSV = "/Users/vinay_kl/Downloads/Test05.csv"
DownloadPath = "/Users/vinay_kl/Downloads/Test05_ffill.csv"
```

```python
with open(dataCSV, 'r') as f:
    reader = csv.reader(f)
    the_whole_file = list(reader)    
    
count = 0
while(count < len(the_whole_file)):
    count = count + 1
    Utterance = the_whole_file[count][2]
    
    headers = {'Content-Type':'application/json'}
    payload = json.dumps ({
        "text": Utterance
    })
    response = requests.request("POST", "http://20.207.84.64:8000/predict", headers=headers, data=payload).json()

    pprint(response["response"]["label"][0]["name"])
```
